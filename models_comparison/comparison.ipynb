```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ["ResNet50", "MobileNetV2", "DenseNet121"]
```

```python
accuracies = [0.9867, 0.9861, 0.9915]
macro_avgs = [0.9876, 0.9870, 0.9915]
weighted_avgs = [0.9873, 0.9870, 0.9916]
```

```python
x = np.arange(len(models))
width = 0.25
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x - width, accuracies, width=width, label='Accuracy')
plt.bar(x, macro_avgs, width=width, label='Macro Avg')
plt.bar(x + width, weighted_avgs, width=width, label='Weighted Avg')

plt.xticks(x, models, fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.ylim(0.98, 1.0)
plt.title("📊 Model Comparison - Accuracy & Averages", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("model_comparison_metrics.png")
plt.show()
```
